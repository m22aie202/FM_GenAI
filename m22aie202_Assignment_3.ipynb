```
import torch
import torch.nn as nn
from transformers import BertTokenizer, BertModel
from torch.optim import AdamW
from datasets import load_dataset
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
from transformers import logging
from tqdm import tqdm

tqdm(disable=True)  # disables widget rendering for saving
logging.set_verbosity_error()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```
dataset = load_dataset("stanfordnlp/sst2")

train_data = dataset["train"].shuffle(seed=42).select(range(10000))
val_size = min(1000, len(dataset["validation"]))
val_data = dataset["validation"].select(range(val_size))
```

```
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

def tokenize(batch):
    return tokenizer(
        batch["sentence"],
        padding="max_length",
        truncation=True,
        max_length=128,
        return_tensors="pt"
    )

train_data = train_data.map(tokenize, batched=True)
val_data = val_data.map(tokenize, batched=True)

train_data.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])
val_data.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])

train_loader = DataLoader(train_data, batch_size=16, shuffle=True)
val_loader = DataLoader(val_data, batch_size=16)
```

```
class Adapter(nn.Module):
    def __init__(self, hidden_size, bottleneck=64):
        super(Adapter, self).__init__()
        self.down = nn.Linear(hidden_size, bottleneck)
        self.relu = nn.ReLU()
        self.up = nn.Linear(bottleneck, hidden_size)

    def forward(self, x):
        return x + self.up(self.relu(self.down(x)))  # residual connection
```

```
class BertWithAdapters(nn.Module):
    def __init__(self):
        super().__init__()
        self.bert = BertModel.from_pretrained("bert-base-uncased")
        hidden_size = self.bert.config.hidden_size

        # base model
        for param in self.bert.parameters():
            param.requires_grad = False

        # adapters to encoder layer
        self.adapters = nn.ModuleList([
            Adapter(hidden_size) for _ in range(len(self.bert.encoder.layer))
        ])

        self.classifier = nn.Linear(hidden_size, 2)

    def forward(self, input_ids, attention_mask):
        outputs = self.bert(
            input_ids=input_ids,
            attention_mask=attention_mask,
            output_hidden_states=True
        )

        hidden_states = outputs.last_hidden_state

        # layer-wise adapters
        for i, adapter in enumerate(self.adapters):
            hidden_states = adapter(hidden_states)

        cls_output = hidden_states[:, 0]
        logits = self.classifier(cls_output)
        return logits
```

```
#load model
model = BertWithAdapters().to(device)
```

```
total_params = sum(p.numel() for p in model.parameters())
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

print(f"total Parameters: {total_params}")
print(f"trainable Parameters: {trainable_params}")
```

```
optimizer = AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=2e-4)
criterion = nn.CrossEntropyLoss()

train_losses = []
val_losses = []

epochs = 5

for epoch in range(epochs):
    model.train()
    total_train_loss = 0

    for batch in train_loader:
        optimizer.zero_grad()

        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["label"].to(device)

        outputs = model(input_ids, attention_mask)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        total_train_loss += loss.item()

    avg_train_loss = total_train_loss / len(train_loader)
    train_losses.append(avg_train_loss)

    # validate
    model.eval()
    total_val_loss = 0

    with torch.no_grad():
        for batch in val_loader:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["label"].to(device)

            outputs = model(input_ids, attention_mask)
            loss = criterion(outputs, labels)

            total_val_loss += loss.item()

    avg_val_loss = total_val_loss / len(val_loader)
    val_losses.append(avg_val_loss)

    print(f"Epoch {epoch+1} | Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f}")
```

```
plt.plot(train_losses, label="Train Loss")
plt.plot(val_losses, label="Validation Loss")
plt.legend()
plt.title("Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

```
def predict(sentence):
    model.eval()
    inputs = tokenizer(
        sentence,
        return_tensors="pt",
        truncation=True,
        padding=True,
        max_length=128
    ).to(device)

    with torch.no_grad():
        outputs = model(inputs["input_ids"], inputs["attention_mask"])
        prediction = torch.argmax(outputs, dim=1).item()

    return "Positive" if prediction == 1 else "Negative"


# examples
print(predict("The movie was absolutely fantastic"))
print(predict("Worst film I have ever seen"))
```

Fully fine-tuning large transformer models like BERT can be expensive and inefficient. Since BERT has over 110 million parameters, updating all of them requires significant GPU memory, longer training time, and higher computational cost. Additionally, for every new task, a separate fully fine-tuned model must be stored, which increases storage requirements and makes scaling to multiple tasks impractical.

To address these challenges, Parameter-Efficient Fine-Tuning (PEFT) introduces small, modular components called adapters inside the transformer layers. Instead of retraining the entire model, the base BERT weights are frozen, and only the adapters and the final classification layer are trained. These adapters use a bottleneck structure to learn task-specific patterns with very few additional parameters.

In this implementation, only about 1% of the total parameters were trainable, which significantly reduced training cost while still achieving good performance. The training results show steady convergence, with validation loss improving from 0.3443 in the first epoch to around 0.3320 by epoch five. This indicates that the adapters were able to effectively learn sentiment-specific features without modifying the core language model.

Overall, adding these modular components inside the transformer makes fine-tuning more efficient, scalable, and practical, especially when working with limited computational resources or multiple downstream tasks.
